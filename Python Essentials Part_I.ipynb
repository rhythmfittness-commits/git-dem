# What is Jupyter Notebook?

Jupyter Notebook is an open-source **web application** that allows data scientists and analysts to create and share documents that combine:
Python code, Mathematical equations,Outputs (tables, numbers), Visualizations (graphs, charts), Images, text.

All in one interactive document.

### Where is Jupyter Notebook Used?

Jupyter Notebook is commonly used for:Data Cleaning & Data Transformation,Numerical & Scientific Computation,Statistical Analysis,Data Visualization,Machine Learning & AI model building,Research, experimentation, and teaching.

### Why Jupyter Notebook? (Key Features)

* Interactive:You can write code, run it, and see output immediately.
* Results are stored:Outputs (tables, graphs, results) remain saved in the notebook, so you can review them later.
* Documentation-friendly:Code + explanation + output in one place → perfect for learning, teaching, and reporting.
* Helpful shortcuts:Many keyboard shortcuts to work faster (Shift+Enter, A, B, etc.).
* Industry standard,Widely used in Data Science, Machine Learning, AI, and Research projects.

### format() for a string

```python
x=10
y=20
z=30
print("value of x",x,"y,z is",y,z)
print("value of x is {},value of y is {},value of z is {}".format(x,y,z))
print(f"value of x is {x},value of y is {y},value of z is{z}")
```

```python
# Method 1: Comma separation in print()
x = 10
y = 20
z = 30

# Method 3: f-string (modern method)
print(f"value of x,y,z is: {x}, {y} and {z}")

# Even shorter examples:
name = "John"
age = 25
print("my",name, "is", age, "years old")  # Method 1
print("{} is {} years old".format(name, age))  # Method 2
print(f"{name} is {age} years old")  # Method 3
```

```python
# Combining string & numbers using format method
item1 = 40
item2 = 55
item3 = 77
"Cost of the three items are {} , {} and {} respectively.".format(item1,item3,item2)
```

# Typecasting

Type Casting is the process of a variable's data type converted into another type in order for an operation to be performed.

* There are 2 ways in which typecasting occurs:

1) Auto Typecasting

2) Forced Typecasting

## Auto Typecasting
* The Python interpreter automatically converts one data type to another without any user involvement.

```python
z=5+False
print(z)
type(z)
```

```python
a=1
b=1+4j
c=a+b
type(c)
```

# Part 1: Python Operators

### 1. Arithmetic Operators
These perform mathematical operations:

```python
# Two numbers for calculation
a = 10
b = 3

# Addition: adding both numbers
print("Addition:", a + b)

# Subtraction: subtracting second number from first
print("Subtraction:", a - b)

# Multiplication: multiplying both numbers
print("Multiplication:", a * b)

# Division: gives answer in decimal
print("Division:", a / b)

# Floor Division: removes decimal part and keeps only whole number
print("Floor Division:", a // b)

# Modulus: gives the remainder after division
print("Remainder (Modulus):", a % b)

# Exponent: a raised to the power b (a to the power b)
print("Power:", a ** b)

# The plus operator can help concatenate string
x='Data'
y ='Science'
z = x + y
print(z)
```

### 2. Comparison (Relational) Operators
These compare values and return True or False:

```python
# Two numbers for comparison
x = 10
y = 5

# Greater than
print("Is  x  greater  than  y :", x > y)

# Less than
print(" Is x less than y :", x < y)

# Equal to
print("Is  x  equal  to    y  :", x == y)

# Not equal to
print("Is  x  not  equal  to  y:", x != y)

# Greater than or equal to
print("Is  x  greater  than  or  equal  to  y :", x >= y)

# Less than or equal to
print("x  less  than  or  equal  to  y :", x <= y)

# Comparison operators on string
a = 'Data'
b = 'Science'
c = 'Data'
print(a == b)
print(a == c)
print(a != b)
```

# 3. Logical Operators
Combine multiple conditions:

```python
# Logical Operators Example

# AND: result is True only if both conditions are True
print("AND operation  :", True and True)
print("AND operation :", True and False)

# OR: result is True if at least one condition is True
print(" OR operation  :", True or False)
print(" OR operation :", False or False)

# NOT: reverses the result
print("NOT operation  :", not True)
print("NOT operation: ", not False)
```

# Part 2: Control Flow Statements

### 1. if Statement
* The _if_ statement evaluates an expression and executes a block of code on the condition that the expression is true.
* In case the expression returns a False, the _if_ block-of-code is ignored.

```python
# Checking today's temperature
t = 20

# If the temperature is more than 30
if t > 30:
    print("It is a hot day")   # Indentation for the block-of-code
print("Independent statement") # Not indented
```

### 2. if-else Statement
* else may be included to account for alternate statements when if condition is false

```python
# if-else structure
age = 16

if age >= 18:
    print("You are an adult") 
else:
    print("You are a minor")

print("This line always executes")  # Outside if-else block
```

* List is create by Square bracket []
* it also follow order.
* it is mutable.

```python
# Creating lists
list1 = []
print(list1,type(list1))
numbers = [1, 2, 3, 4, 5]
fruits = ["apple", "banana", "cherry"]
mixed = [10, "hello", 3.14, True]  # Can hold different types
nested = [[1, 2], [3, 4], [5, 6]]  # List of lists

# Accessing elements
print(fruits[0])      # "apple" - first element
print(fruits[-1])     # "cherry" - last element
print(fruits[1:3])    # ["banana", "cherry"] - slicing
print(fruits[1:3:2])

# Modifying lists
fruits.append("orange")      # Add to end
fruits.insert(1, "mango")    # Insert at specific position
fruits[0] = "kiwi"           # Change element, Replace element
fruits.remove("banana")      # Remove by value
popped = fruits.pop(2)       # Remove by index and get value
fruits.pop()                 # removes last element by default
del fruits[0]                # Delete by index
fruits.clear()               # Remove all elements

# List operations
list1 = [1, 2, 3]
list2 = [4, 5, 6]

combined = list1 + list2      # Concatenation
repeated = list1 * 3          # Repetition
length = len(list1)           # Length

# Membership check: True
print(3 in list1)             

# Sorting
numbers = [5, 2, 8, 1, 9]
numbers.sort()                # Ascending (modifies original)
print(numbers)                # [1, 2, 5, 8, 9]
list1[::-1] 
sorted_numbers = sorted(numbers, reverse=True)  # Descending (new list)
```

### 2. Tuples - Ordered, Immutable Collection
* Tuple is similar to a list except that a tuple is immutable - we cannot change the elements of a tuple once assigned.
* When we do not want to change the data over time, tuple is a preferred data type.
* Iterating over the elements of a tuple is faster compared to iterating over a list.

```python
# Creating tuples
empty_tuple = ()
single_item = (5,)  # Comma is important!
colors = ("red", "green", "blue")
scores = (75,38,77,90,75,86,60,92,75,36,82,84,75,75)

# Accessing (same as lists but cannot modify)
print(colors[0])      # "red"
print(colors[-1])     # "blue"
print(colors[1:])     # ("green", "blue")

scores.count(75) # no. of occurrences of a given item through the tuple
scores.index(75) # locate the position of an item

# Why use tuples?
# 1. Faster than lists
# 2. Safer (can't be changed accidentally)
# 3. Can be used as dictionary keys (lists cannot)
# 4. Return multiple values from functions

# Packing and unpacking
point = (5, 10)           # Packing
x, y = point              # Unpacking
print(f"x: {x}, y: {y}")

# Swapping variables
a, b = 10, 20
a, b = b, a  # Swapped using tuple unpacking
print(f"a: {a}, b: {b}")
```

### 3. Sets - Unordered, Unique Elements

```python
# Creating sets
empty_set = set()  # NOT {} - that's for dictionary!
numbers = {1, 2, 3, 4, 5}
duplicate_set = {1, 2, 2, 3, 3, 3}  # Becomes {1, 2, 3}

# Set operations
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

print(set1.union(set2))        # {1, 2, 3, 4, 5, 6, 7, 8}
print(set1.intersection(set2)) # {4, 5}
print(set1.difference(set2))   # {1, 2, 3}
print(set2.difference(set1))   # {6, 7, 8}
print(set1.symmetric_difference(set2))  # {1, 2, 3, 6, 7, 8}

# Membership is VERY fast in sets
fruits_set = {"apple", "banana", "cherry"}
print("banana" in fruits_set)  # True 

# Modifying sets
fruits_set.add("orange")
fruits_set.update(["mango", "grape"])
fruits_set.remove("apple")     # Error if not exists
fruits_set.discard("kiwi")     # No error if not exists
popped = fruits_set.pop()      # Removes random element
```

### 4. Dictionaries - Key-Value Pairs

```python
# Creating dictionaries
empty_dict = {}
student = {
    "name": "John",
    "age": 20,
    "courses": ["Math", "Science"]
}

# Accessing values
print(student["name"])          # Direct access
print(student.get("age"))       # Safe access (returns None if key doesn't exist)
print(student.get("grade", "N/A"))  # Default value

# Adding/Modifying
student["grade"] = "A"          # Add new key
student["age"] = 21             # Modify existing
student.update({"city": "NYC", "age": 22})  # Multiple updates

# Removing
removed = student.pop("grade")  # Remove and get value
del student["city"]             # Remove key
student.clear()                 # Remove all

# Dictionary views
print(student.keys())           # Keys
print(student.values())         # Values
print(student.items())          # Key-value pairs

# # Iterating through dictionary
#  for key in student:
#      print(f"{key}: {student[key]}")

# # for key, value in student.items():
# #     print(f"{key}: {value}")
```
